### 우선 area_data와 road_data merge

### merged_area_road 데이터와 estimatedrevenue_data 데이터 merge()

### merged_area_road_revenue 데이터와 store_data merge()

```python
# 모든 데이터프레임의 컬럼명을 세트로 변환하여 리스트에 저장
dataframes = [merged_area_road_revenue, store_data]
column_sets = [set(df.columns) for df in dataframes]

# 모든 데이터프레임의 컬럼명의 교집합(공통된 컬럼)을 추출
common_columns = set.intersection(*column_sets)

# 공통된 컬럼 출력
print("공통된 컬럼:", common_columns)
```

```python
merged_area_road_revenue_store = pd.merge(merged_area_road_revenue, store_data, on=['기준_년분기_코드', '상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명', '서비스_업종_코드_명', '서비스_업종_코드'], how='outer')
print(list(merged_area_road_revenue_store.columns))
merged_area_road_revenue_store.head()
```

### merged_area_road_revenue_store 데이터와 나머지 데이터 merge()

```python
# 모든 데이터프레임의 컬럼명을 세트로 변환하여 리스트에 저장
dataframes = [merged_area_road_revenue_store, workforce_data, public_data, resident_data, income_data]
column_sets = [set(df.columns) for df in dataframes]

# 모든 데이터프레임의 컬럼명의 교집합(공통된 컬럼)을 추출
common_columns = set.intersection(*column_sets)

# 공통된 컬럼 출력
print("공통된 컬럼:", common_columns)
```

```python
merged_df = pd.merge(merged_area_road_revenue_store, workforce_data, on=['기준_년분기_코드', '상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명'], how='outer')
merged_df = pd.merge(merged_df, public_data, on=['기준_년분기_코드', '상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명'], how='outer')
merged_df = pd.merge(merged_df, resident_data, on=['기준_년분기_코드', '상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명'], how='outer')
merged_df = pd.merge(merged_df, income_data, on=['기준_년분기_코드', '상권_구분_코드', '상권_구분_코드_명', '상권_코드', '상권_코드_명'], how='outer')
```
